```python
import pandas as pd
from collections import defaultdict
```

```python
review = pd.read_csv('review.csv')
```

```python
truePositives = defaultdict(int)
falsePositives = defaultdict(int)

for _, row in review.iterrows():
    condition = row['condition group']
    classification = row['classification']
    if classification == 'true positive':
        truePositives[condition] += 1
    elif classification == 'false positive':
        falsePositives[condition] += 1

ppvValues = {}
for condition in set(review['condition group']):
    tp = truePositives[condition]
    fp = falsePositives[condition]
    ppv = tp / (tp + fp) if (tp + fp) > 0 else 0
    print(f'condition: {condition}')
    print(f'  true positives: {tp}')
    print(f'  false positives: {fp}')
    print(f'  ppv: {ppv:.2f}')
    ppvValues[condition] = ppv
```

```python
averagePPV = sum(ppvValues.values()) / len(ppvValues)
print(f'average ppv across all groups: {averagePPV:.2f}')
```
